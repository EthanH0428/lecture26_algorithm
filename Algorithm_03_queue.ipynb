```python
# 정수형 큐 연산 파이썬 구현 실습

from collections import deque
queue = deque()

def enqueue(data):
    queue.append(data)

def dequeue():
    if queue:
        data = queue.popleft()
    else:
        data = -1
    return data

def getFront():
    if queue:
        return queue[0]
    else:
        return -1

print('시작')

while True:
    menu = int(input('>>'))

    if menu == 0:
        break

    elif menu == 1:
        data = int(input('데이터:'))
        enqueue(data)

    elif menu == 2:
        data = dequeue()
        if data != -1:
            print('가져온 데이터:', data)
        else:
            print('Queue가 비어 있습니다')
    
    elif menu == 3:
        data = getFront()
        if data != -1:
            print(data)
        else:
            print('Queue가 비어 있습니다')
    print(queue)

print('종료')
```
