### 약수 구하기

```python
def get_divisors_list(n):
    divisors = []
    for i in range(1, n + 1):
        if n % i == 0:
            divisors.append(i)
    return divisors

print(get_divisors_list(12)) 
```

### 공약수 구하기

```python
def getcd(m, n):
    if m > n: m, n = n, m
    cdivisor_list = []
    for num in range(1, m + 1):
        if m % num == 0 and n % num == 0:
            cdivisor_list.append(num)
            
    return cdivisor_list
print(getcd(12, 18))
```

###  최대 공약수 구하기

```python
# 최대공약수
def getGCD(M, N):
    if M > N: M, N = N, M # M이 작은 숫자가 되도록 조정

    for num in range(M, 0, -1):
        if M % num == 0 and N % num ==0:
            return num  # 최대공약수

print(getGCD(12, 18))
```

###  최소 공배수 구하기

```python
# 최소공배수
def getLCM(M, N):
    if M > N: M, N = N, M # M이 작은 숫자가 되도록 조정

    # 큰 숫자 N의 배수를 구하면서, M의 배수인지 확인
    # 최소 공배수 -> 공배수 중 가장 작은 애 -> 처음 만나는 애
    i = 1
    while True:
        multiple = N * i
        if multiple % M == 0: # M의 배수인지 확
            return multiple
        i += 1
        
print(getLCM(12, 18))
```

```python
def is_prime_number(N):
    if N == 1: return False

    is_prime = True
    for i in range(2, N):
        if N % i == 0:
            is_prime = False
            break

    return is_prime
print(is_prime_number(2))
print(is_prime_number(1))
print(is_prime_number(3))
print(is_prime_number(5))
print(is_prime_number(6))
print(is_prime_number(7))
```

### 소수(Prime number)

```python
# 알고리즘 개선하기(1)
def is_prime_number(n):
    if n == 1: return False
    if n == 2: return True
        
# 2의 배수인지 확인 (짝수 제외)
    if n % 2 == 0:
        return False
    
# 3부터 N-1까지의 홀수에 대해서만 확인
    for i in range(3, n, 2):
        if n % i == 0:
            return False
    return True

print(is_prime_number(17))
```

```python
# 알고리즘 개선하기(2)
def is_prime_2(n):
    if n < 2: return False
    if n == 2: return True
    
# 2부터 N/2까지의 수로만 나누어지는지 확인
    for i in range(2, n // 2 + 1):
        if n % i == 0:
            return False
    return True

print(is_prime_2(20))
```
